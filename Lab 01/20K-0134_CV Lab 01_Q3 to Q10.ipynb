## Q3

```python
import cv2
import matplotlib.pyplot as plt
image = cv2.imread("image.jpg")

plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Original Image")
plt.axis("off")

gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.subplot(1, 3, 2)
plt.imshow(gray_image, cmap='gray')
plt.title("Grayscale Image")
plt.axis("off")

new_size = (300, 100)
resized_image = cv2.resize(image, new_size)
plt.subplot(1, 3, 3)
plt.imshow(cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB))
plt.title("Resized Image")
plt.axis("off")

plt.tight_layout()
plt.show()
```

## Q4

```python
import numpy as np
image_draw = np.zeros((300, 400, 3), dtype=np.uint8)
cv2.rectangle(image_draw, (50, 50), (200, 150), (0, 0, 255), -1)
cv2.circle(image_draw, (300, 200), 50, (0, 255, 0), -1)
cv2.imshow("Drawn Shapes", image_draw)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Q5

```python
blurred_image = cv2.GaussianBlur(image, (5, 5), 0)

plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(blurred_image, cv2.COLOR_BGR2RGB))
plt.title("Blurred Image")
plt.axis("off")

roi = image[100:300, 150:350]
plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(roi, cv2.COLOR_BGR2RGB))
plt.title("Cropped Image")
plt.axis("off")

plt.tight_layout()
plt.show()
```

## Q6

```python
target_width = 800
target_height = 600

plt.figure(figsize=(12, 4))
resized_image = cv2.resize(image, (target_width, target_height))
text = "This is a Nature's photograph"
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(resized_image, text, (50, 150), font, 1.5, (0, 0, 255), 2)
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB))
plt.title("Text on Image")
plt.axis("off")

blank_image = np.zeros((target_height, target_width, 3), dtype=np.uint8)
cv2.putText(blank_image, text, (50, 150), font, 1.5, (0, 0, 255), 2)
plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(blank_image, cv2.COLOR_BGR2RGB))
plt.title("Text on Blank Image")
plt.axis("off")

plt.tight_layout()
plt.show()
```

## Q7

```python
import pandas as pd
pixel_values = gray_image.reshape(-1)
image_df = pd.DataFrame({"Pixel_Value": pixel_values})
print("Basic Statistics of Image Pixel Values:")
print(image_df.describe())

rotation_matrix = cv2.getRotationMatrix2D((image.shape[1] / 2, image.shape[0] / 2), 60, 1)
rotated_image = cv2.warpAffine(image, rotation_matrix, (image.shape[1], image.shape[0]))
cv2.imshow("Rotated Image", rotated_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Q8

```python
image1 = cv2.imread("image1.jpg")
image1 = cv2.resize(image1, (image.shape[1], image.shape[0]))
alpha = 0.5
beta = 0.5
gamma = 0
blended_image = cv2.addWeighted(image, alpha, image1, beta, gamma)
gray_blended_image = cv2.cvtColor(blended_image, cv2.COLOR_BGR2GRAY)
equalized_image = cv2.equalizeHist(gray_blended_image)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(blended_image, cv2.COLOR_BGR2RGB))
plt.title("Blended Image")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(equalized_image, cmap="gray")
plt.title("Enhanced Image")
plt.axis("off")
plt.show()
```

## Q9

```python
height, width, _ = image.shape
mask = np.zeros((height, width), dtype=np.uint8)
cv2.rectangle(mask, (100, 100), (400, 300), 255, -1)

bitwise_and = cv2.bitwise_and(image, image, mask=mask)
bitwise_or = cv2.bitwise_or(image, image, mask=mask)
bitwise_xor = cv2.bitwise_xor(image, image, mask=mask)
bitwise_not = cv2.bitwise_not(image, mask=mask)

plt.figure(figsize=(12, 8))
plt.subplot(2, 3, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Original Image")
plt.axis("off")

plt.subplot(2, 3, 2)
plt.imshow(mask, cmap='gray')
plt.title("Mask")
plt.axis("off")

plt.subplot(2, 3, 3)
plt.imshow(cv2.cvtColor(bitwise_and, cv2.COLOR_BGR2RGB))
plt.title("Bitwise AND")
plt.axis("off")

plt.subplot(2, 3, 4)
plt.imshow(cv2.cvtColor(bitwise_or, cv2.COLOR_BGR2RGB))
plt.title("Bitwise OR")
plt.axis("off")

plt.subplot(2, 3, 5)
plt.imshow(cv2.cvtColor(bitwise_xor, cv2.COLOR_BGR2RGB))
plt.title("Bitwise XOR")
plt.axis("off")

plt.subplot(2, 3, 6)
plt.imshow(cv2.cvtColor(bitwise_not, cv2.COLOR_BGR2RGB))
plt.title("Bitwise NOT")
plt.axis("off")

plt.tight_layout()
plt.show()
```

## Q10

```python
statistics = image_df.describe()

height, width, _ = image.shape
mask = np.zeros((height, width), dtype=np.uint8)
cv2.rectangle(mask, (100, 100), (400, 300), 255, -1)
masked_image = cv2.bitwise_and(image, image, mask=mask)

plt.figure(figsize=(12, 6))
plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Original Image")
plt.axis("off")

plt.subplot(1, 3, 2)
plt.imshow(cv2.cvtColor(masked_image, cv2.COLOR_BGR2RGB))
plt.title("Masked Image")
plt.axis("off")

plt.subplot(1, 3, 3)
plt.axis("off")
plt.table(cellText=statistics.values, colLabels=statistics.columns, cellLoc='center', loc='center')

plt.tight_layout()
plt.show()
```
