```python
name = []
price = []
qty = []
x=10
while x!=0:
    x = int(input("-> Enter 1 to add items.\n-> Enter 2 to remove items.\n-> Enter 3 to display current list.\n-> Enter 0 to exit.\n"))
    if x == 1:
        prod = str(input("Enter Product Name: "))
        name.append(prod)
        cost = float(input("Enter Product Price: "))
        price.append(cost)
        quan = int(input("Enter Product Quantity: "))
        qty.append(quan)
        print("New Product Added!!")
        exit(0)
    elif x == 2:
        if len(name) < 0:
            print("List is Empty!!")
            exit(0)
        else:
            prod = str(input("Enter Name of the Product to remove: "))
            for i in range(len(name)):
                if name[i] == prod:
                    name.pop(i)
                    price.pop(i)
                    qty.pop(i)
                    print("Successful Deletion of Product!!")
                    exit(0)
            print("No such Product Exists!!")
            exit(0)
    elif x == 3:
        if len(name) < 0:
            print("list is Empty!!")
            exit(0)
        else:
            print("Product\tPrice\tQuantity")
            for i in range(len(name)):
                print(name[i], "\t", price[i], "\t", qty[i], "\n")
            exit(0)
    elif x == 0:
        print("You Have Exited!!")
    else:
        print("No such Option!!")
        exit(0)
```

```python
student_records = {}
x = 10
while x != 0:
    x = int(input("-> Enter 1 to add student record.\n-> Enter 2 to update grade.\n-> Enter 3 to view records.\n-> Enter 0 to exit\n"))

    if x == 1:
        stud_id = str(input("Enter Student ID: "))
        if stud_id in student_records:
            print("Student with ID", stud_id, "already exists.")
        else:
            name = str(input("Enter Student Name: "))
            grade = str(input("Enter Student Grade: "))
            student_records[stud_id] = {'name': name, 'grade': grade}
            print("Student with ID ", stud_id, " has been added.")

    elif x == 2:
        stud_id = str(input("Enter Student ID: "))
        if stud_id in student_records:
            new_grade = str(input("Enter New Grade: "))
            student_records[stud_id]['grade'] = new_grade
            print("Grade for student ", student_records[stud_id]['name'], " with ID ", stud_id, " has been updated to ", new_grade, ".")
        else:
            print("Student with ID ", stud_id, " does not exist.")

    elif x == 3:
        if len(student_records) == 0:
            print("No student records found.")
        else:
            print("Student Records:")
            for stud_id, stud_info in student_records.items():
                print("ID: ", stud_id)
                print("Name: ", stud_info['name'])
                print("Grade: ", stud_info['grade'])

    elif x == 0:
        print("You Are Exiting Student Record System!")
        exit(0)

    else:
        print("Invalid choice!")
```
