```python
image_path = 'sample_001.png'
image = plt.imread(image_path)
fig, ax = plt.subplots(1)
ax.imshow(image)

for _, row in train[train.filename == image_path].iterrows():
    xmin, xmax, ymin, ymax = row.x1, row.x2, row.y1, row.y2
    width, height = xmax - xmin, ymax - ymin

    if row['class'] == 'stop':
        edgecolor = 'r'
        label = 'stop'
    elif row['class'] == 'keepRight':
        edgecolor = 'b'
        label = 'keepRight'
    elif row['class'] == 'speedLimit35':
        edgecolor = 'g'
        label = 'speedLimit35'
        
    rect = Rectangle((xmin, ymin), width, height, linewidth=1, edgecolor=edgecolor, facecolor='none', label=label)
    ax.add_patch(rect)

plt.legend()
plt.show()
```
